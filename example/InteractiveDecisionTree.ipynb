## Use DecisionTreeClassifier your own
for example: use wine datasets  

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_wine
from sklearn.tree import DecisionTreeClassifier

wine = load_wine()
X = pd.DataFrame(wine.data, columns=wine.feature_names)
y = pd.Series(wine.target, name='target')
```

```python
clf = DecisionTreeClassifier(min_samples_leaf=.1, max_depth=3)
clf.fit(X, y)
dtree = dtreeplt(model=clf, feature_names=wine.feature_names, target_names=wine.target_names, X=X, y=y, eval=True)
dtree.view(interactive=True)
```
